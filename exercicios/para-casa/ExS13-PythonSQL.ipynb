Import dos códigos

```python
import sqlite3
import csv
```

Criar a conexão com o banco de dados e o cursor

```python
banco = sqlite3.connect('dadosConsumo.db') 
cursor = banco.cursor()
```

Pedir ao cursor pra criar uma tabela no banco de dados com tais colunas e aplicar os métodos da lib CSV pra abrir e ler o arquivo

```python
table = """CREATE TABLE IF NOT EXISTS dadosConsumo(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    customerID INTEGER NOT NULL,
    age INTEGER NOT NULL,
    gender TEXT NOT NULL,
    itemPurchased TEXT NOT NULL,
    category TEXT NOT NULL,
    purchaseAmountUSD INTEGER NOT NULL,
    location TEXT NOT NULL,
    size TEXT NOT NULL,
    color TEXT NOT NULL,
    season TEXT NOT NULL,
    reviewRating REAL NOT NULL,
    subscriptionStatus TEXT NOT NULL,
    shippingType TEXT NOT NULL,
    discountApplied TEXT NOT NULL,
    promoCodeUsed TEXT NOT NULL,
    previousPurchases INTEGER NOT NULL,
    paymentMethod TEXT NOT NULL,
    frequencyOfPurchases TEXT NOT NULL
)"""

cursor.execute(table)

file = open("shopping_trends_updated.csv")

conteudo = csv.reader(file)

next(conteudo)
```

Métodos do sqlite pra fazer o INSERT do conteúdo do CSV na tabela e o SELECT pra consultar as informações

```python
inserir_conteudo = '''INSERT INTO dadosConsumo (customerID, age, gender, itemPurchased, category, purchaseAmountUSD, location, size, color, season, reviewRating, subscriptionStatus, shippingType, discountApplied, promoCodeUsed, previousPurchases, paymentMethod, frequencyOfPurchases
) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)'''

cursor.executemany(inserir_conteudo, conteudo)

selecionar_tudo = '''SELECT * FROM dadosConsumo'''

entradas = cursor.execute(selecionar_tudo).fetchall()

for linha in entradas:
    print(linha)
```

Commit e fechamento da conexão

```python
banco.commit()

banco.close()
```
